```python
#필요한 라이브러리 import
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#데이터 로드
df_continue = pd.read_csv('/Users/ujong-yun/Documents/학부연구생/100%_load_continued_all.csv', encoding='euc-kr')
df_onoff = pd.read_csv('/Users/ujong-yun/Documents/학부연구생/100%_load_daily_on-off_all.csv', encoding='euc-kr')

#데이터 shape 확인
print(df_continue.shape)
print(df_onoff.shape)


#데이터 칼럼별 시각화 함수
def Data_Visualization(df):
    fig, axs = plt.subplots(6, 2, figsize=(10, 10))
    j = 0
    for i in range((int(len(df.columns)))):
        axs[j][i % 2].plot(df.index, df.iloc[:, i])  # 수정된 부분
        axs[j][i % 2].set_xlabel('Count')
        axs[j][i % 2].set_ylabel(df.columns[i])
        if i % 2 == 1:
            j += 1
    plt.tight_layout()
    plt.show()


print(Data_Visualization(df_continue))
print(Data_Visualization(df_onoff))

```
