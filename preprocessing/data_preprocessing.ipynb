```python
#필요한 라이브러리 import
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#데이터 로드
df_continue = pd.read_csv('/Users/ujong-yun/Documents/학부연구생/100%_load_continued_all.csv', encoding='euc-kr')
df_onoff = pd.read_csv('/Users/ujong-yun/Documents/학부연구생/100%_load_daily_on-off_all.csv', encoding='euc-kr')

#데이터 shape 확인
print(df_continue.shape)
print(df_onoff.shape)

print(len(df_continue.index))
#데이터 칼럼별 시각화 함수
def Data_Visualization(df):
    fig, axs = plt.subplots(1, len(df.columns), figsize = (10, 10), constrained_layout = True)
    for i in range(1, len(df.columns)):
        axs[i-1].plot(df.index, df.iloc[:, [i]])
        axs[i-1].set_xlabel('Count')
        axs[i-1].set_ylabel(df.columns[i])
        plt.show()
```

```python
print(Data_Visualization(df_continue))
```
